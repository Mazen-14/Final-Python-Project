```python
import csv
```

```python
class Employee:
    def __init__(self, id, name, position, salary, email):
        self.id = id
        self.name = name
        self.position = position
        self.salary = salary
        self.email = email
    def update_id(self, new_id):
        self.id = new_id
    def update_name(self, new_name):
        self.name = new_name
    def update_position(self, new_position):
        self.position = new_position
    def update_salary(self, new_salary):
        self.salary = new_salary
    def update_email(self, new_email):
        self.email = new_email
        
    def display(self):
        print("------------------------------------------------------------------------------------------")
        print(f"ID: {self.id} Name: {self.name} Position: {self.position} Salary: {self.salary} Email: {self.email}")
        

```

```python
class EmployeeManager:
    def __init__(self):
        self.filename = "emp.csv"
        self.emp = self.load_csv()

    def load_csv(self):
        emp = []
        with open(self.filename, mode='r', newline='', encoding="utf-8") as file:
                reader = csv.reader(file)
                next(reader) 
                for row in reader:
                    id, name, position, salary, email = row
                    emp.append(Employee(id, name, position, float(salary), email))
        return emp

    def save_csv(self):
        
            with open(self.filename, mode='w', newline='', encoding="utf-8") as file:
                writer = csv.writer(file)
                writer.writerow(["ID", "Name", "Position", "Salary", "Email"])
                for emp in self.emp:
                    writer.writerow([emp.id, emp.name, emp.position, emp.salary, emp.email])
    
    def get_by_id(self, emp_id):
        for emp in self.emp:
            if emp.id == emp_id:
                return emp

    def add_employee(self, id, name, position, salary, email):
        if self.get_by_id(id):
            print("Employee already exists")
            return
        new_employee = Employee(id, name, position, salary, email)
        self.emp.append(new_employee)
        self.save_csv()
        

    def update_employee(self, id, field, new):
        emp = self.get_by_id(id)
        if emp:
            if field == "name":
                emp.update_name(new)
            elif field == "position":
                emp.update_position(new)
            elif field == "salary":
                emp.update_salary(float(new))
            elif field == "email":
                emp.update_email(new)
            self.save_csv()
            print(f"{field} is updated")
        else:
            print("Not found")

    def delete_employee(self, id):
        emp = self.get_by_id(id)
        if emp:
            self.emp.remove(emp)
            self.save_csv()
            print(f"Employee with ID {id} is deleted")
        else:
            print("Not found")

    def search(self, emp_id):
        emp = self.get_by_id(emp_id)
        if emp:
            emp.display()
            print("------------------------------------------------------------------------------------------")
        else:
            print("------------------------------------------------------------------------------------------")
            print("Not found")
            print("------------------------------------------------------------------------------------------")

    def list(self):
        for emp in self.emp:
            emp.display()
            print("------------------------------------------------------------------------------------------")

    def valid_salary(self, salary):
        is_number = isinstance(salary, (int, float))
        is_post = salary >= 0
        return is_number and is_post
```

```python
manager = EmployeeManager()
print("\n------------------------------------------------------------------------------------------")
print("-----------------------------Welcome to Mazen Project-------------------------------------")
print("------------------------------------------------------------------------------------------")
print("\nPlease choose a number :)")
print("------------------------------------------------------------------------------------------")
print("Add Employee---1")
print("Update Employee---2")
print("Delete Employee---3")
print("Search Employee---4")
print("List All Employees---5")
print("Exit---6")
choose = input("Choose an option: ")

if choose == '1':
    id = input("ID: ")
    name = input("Name: ")
    position = input("Position: ")
    salary = float(input("Salary: "))
    email = input("Email: ")
    if manager.valid_salary(salary):
        manager.add_employee(id, name, position, salary, email)
        print("------------------------------------------------------------------------------------------")
        print(f"Employee with ID {id} is added")
        print("------------------------------------------------------------------------------------------")
    else:
        print("------------------------------------------------------------------------------------------")
        print("Wrong Salary")
        print("------------------------------------------------------------------------------------------")
        
elif choose == '2':
        id = input("ID: ")
        field = input("Field to update \n#name, position, salary, email)#: ")
        new_value = input(f"Enter new {field}: ")
        manager.update_employee(id, field, new_value.lower())
        print("------------------------------------------------------------------------------------------")
        print(f"Employee with ID {id} {field} is updated")
        print("------------------------------------------------------------------------------------------")

elif choose == '3':
        id = input("ID to delete: ")
        manager.delete_employee(id)
        print("------------------------------------------------------------------------------------------")
        print(f"Employee with ID {id} is deleted")
        print("------------------------------------------------------------------------------------------")
        
elif choose == '4':
        id = input("ID: ")
        manager.search(id)

elif choose == '5':
        manager.list()

elif choose == '6':
        print("------------------------------------------------------------------------------------------")
        print("Good bye Eng Baraa :)")
        print("------------------------------------------------------------------------------------------")
else:
        print("------------------------------------------------------------------------------------------")
        print("Invalid number\nEnter again")
        print("------------------------------------------------------------------------------------------")
```
